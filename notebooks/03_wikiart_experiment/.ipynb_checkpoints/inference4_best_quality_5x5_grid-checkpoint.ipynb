# Experiment 3 – WikiArt: Best Quality 5×5 Generation Grid

Generate a **5 styles × 5 samples** matrix using the optimal hyperparameters found during evaluation:

- **Guidance scale:** `w = 1.0` (best composite + CLIP score)
- **Inference steps:** `500` (default high-quality setting)

Each row corresponds to one art style; each column is an independently generated sample
(unique seed), showing intra-style diversity at peak quality.

```python
# ─── Experiment hyperparameters ───────────────────────────────────────────────
GUIDANCE_SCALE     = 1.0   # best composite + CLIP score from guidance scale search
NUM_INFERENCE_STEPS = 100  # high-quality setting (default from INFERENCE_CONFIG)
NUM_SAMPLES        = 5     # columns in the grid (samples per style)
BASE_SEED          = 42    # column i uses seed BASE_SEED + i for reproducibility

# 5 visually diverse, well-known styles
SELECTED_STYLES = [
    "Impressionism",
    "Baroque",
    "Cubism",
    "Pop_Art",
    "Ukiyo_e",
]

print(f"Guidance scale  : {GUIDANCE_SCALE}")
print(f"Inference steps : {NUM_INFERENCE_STEPS}")
print(f"Samples/style   : {NUM_SAMPLES}")
print(f"Seed range      : {BASE_SEED} – {BASE_SEED + NUM_SAMPLES - 1}")
print(f"\nSelected styles :")
for s in SELECTED_STYLES:
    print(f"  • {s.replace('_', ' ')}")
```

## 2. Load Models

```python
# Device setup
if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"Using device: {device}  ({torch.cuda.get_device_name(0)})")
    print(f"VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
else:
    device = torch.device("cpu")
    print(f"Using device: {device}  (CUDA not available)")
```

```python
# Load trained WikiArt UNet from the latest checkpoint
from models.custom_unet_wikiart import CustomUNet2DConditionModelWikiArt

checkpoint_path = get_latest_wikiart_unet_checkpoint()
print(f"Loading checkpoint: {checkpoint_path}")

checkpoint = torch.load(str(checkpoint_path), map_location=device)

unet = CustomUNet2DConditionModelWikiArt().to(device)

if "unet_state_dict" in checkpoint:
    unet.load_state_dict(checkpoint["unet_state_dict"])
elif "model_state_dict" in checkpoint:
    unet.load_state_dict(checkpoint["model_state_dict"])
elif "state_dict" in checkpoint:
    unet.load_state_dict(checkpoint["state_dict"])
else:
    unet.load_state_dict(checkpoint)

unet.eval()

epoch = checkpoint.get("epoch", "unknown") if isinstance(checkpoint, dict) else "unknown"
print(f"✓ Loaded UNet from epoch {epoch}  ({unet.get_num_parameters():,} parameters)")
```

```python
# Load CLIP text encoder and tokenizer (frozen)
print("Loading CLIP text encoder…")
text_encoder = CLIPTextModel.from_pretrained(CLIP_MODEL_NAME).to(device)
text_encoder.eval()

tokenizer = CLIPTokenizer.from_pretrained(CLIP_MODEL_NAME)
print(f"✓ Loaded CLIP  ({CLIP_MODEL_NAME})")
```

## 3. Generation Function

```python
@torch.no_grad()
def generate_image(
    prompt: str,
    guidance_scale: float = 1.0,
    num_inference_steps: int = 500,
    seed: int = 42,
) -> np.ndarray:
    """
    Generate a single 128×128 WikiArt image.

    Args:
        prompt:              Text prompt (e.g. 'A painting in the style of Impressionism')
        guidance_scale:      CFG weight w
        num_inference_steps: Number of DDPM denoising steps
        seed:                RNG seed for reproducibility

    Returns:
        HWC uint8 numpy array in [0, 255]
    """
    # Encode text
    text_input = tokenizer(
        [prompt],
        padding="max_length",
        max_length=TOKENIZER_MAX_LENGTH,
        truncation=True,
        return_tensors="pt",
    )
    text_emb = text_encoder(text_input.input_ids.to(device))[0]

    # Unconditional embedding for CFG
    uncond_input = tokenizer(
        [""],
        padding="max_length",
        max_length=TOKENIZER_MAX_LENGTH,
        return_tensors="pt",
    )
    uncond_emb = text_encoder(uncond_input.input_ids.to(device))[0]

    # Stack: [uncond, cond]
    combined_emb = torch.cat([uncond_emb, text_emb])

    # Seeded noise initialisation
    generator = torch.Generator(device=device).manual_seed(seed)
    latents = torch.randn((1, 3, 128, 128), generator=generator, device=device)

    # Scheduler
    scheduler = DDPMScheduler(
        beta_schedule=INFERENCE_CONFIG["beta_schedule"],
        num_train_timesteps=INFERENCE_CONFIG["num_train_timesteps"],
    )
    scheduler.set_timesteps(num_inference_steps)

    # Denoising loop
    for t in scheduler.timesteps:
        latent_input = torch.cat([latents] * 2)
        latent_input = scheduler.scale_model_input(latent_input, t)

        noise_pred = unet(latent_input, t, encoder_hidden_states=combined_emb).sample
        noise_pred_uncond, noise_pred_text = noise_pred.chunk(2)
        noise_pred = noise_pred_uncond + guidance_scale * (noise_pred_text - noise_pred_uncond)

        latents = scheduler.step(noise_pred, t, latents).prev_sample

    # Denormalize [-1, 1] → [0, 255] uint8
    img = (latents[0] / 2 + 0.5).clamp(0, 1)
    img = (img.cpu().numpy() * 255).astype(np.uint8)
    img = np.transpose(img, (1, 2, 0))  # CHW → HWC
    return img


print("Generation function defined")
```

## 4. Generate 5×5 Grid

Generate 5 samples for each of the 5 selected styles (25 images total).
Each column uses a different random seed so samples are independent.

```python
# Generate all 5 × 5 = 25 images
# grid[row][col] → HWC uint8 array
grid = []

total = len(SELECTED_STYLES) * NUM_SAMPLES
done  = 0

for style_name in SELECTED_STYLES:
    style_display = style_name.replace("_", " ")
    prompt = EXPERIMENT_3_CONFIG["prompt_template"].format(style_name=style_display)
    row_images = []

    for sample_idx in range(NUM_SAMPLES):
        seed = BASE_SEED + sample_idx
        img  = generate_image(
            prompt,
            guidance_scale=GUIDANCE_SCALE,
            num_inference_steps=NUM_INFERENCE_STEPS,
            seed=seed,
        )
        row_images.append(img)
        done += 1
        print(f"  [{done:2d}/{total}]  {style_display}  (seed={seed})")

    grid.append(row_images)

print("\n✓ All images generated")
```

## 5. Plot the 5×5 Matrix

```python
N_STYLES  = len(SELECTED_STYLES)   # rows
N_SAMPLES = NUM_SAMPLES            # columns

IMG_SIZE  = 2.5   # inches per cell
PAD_LEFT  = 1.8   # extra left margin for row labels

fig, axes = plt.subplots(
    N_STYLES, N_SAMPLES,
    figsize=(N_SAMPLES * IMG_SIZE + PAD_LEFT, N_STYLES * IMG_SIZE + 0.6),
    gridspec_kw={"hspace": 0.08, "wspace": 0.04},
)

for row, style_name in enumerate(SELECTED_STYLES):
    style_display = style_name.replace("_", " ")
    for col in range(N_SAMPLES):
        ax = axes[row, col]
        ax.imshow(grid[row][col])
        ax.set_xticks([])
        ax.set_yticks([])

        sample_num = col + 1
        sample_seed = BASE_SEED + col

        # Show sample number + seed on every tile for exact reproducibility
        ax.text(
            0.02, 0.98,
            f"S{sample_num} | seed={sample_seed}",
            transform=ax.transAxes,
            ha="left",
            va="top",
            fontsize=7,
            color="white",
            bbox=dict(facecolor="black", alpha=0.55, pad=1.5, edgecolor="none"),
        )

        # Row labels (left-most column only)
        if col == 0:
            ax.set_ylabel(style_display, fontsize=9, labelpad=6, rotation=0,
                          ha="right", va="center")

# Overall title
fig.suptitle(
    f"WikiArt Generation Samples  ·  w = {GUIDANCE_SCALE}  ·  {NUM_INFERENCE_STEPS} steps",
    fontsize=13, y=1.01,
)

plt.tight_layout()
plt.show()

print("Grid displayed")
```
